```python
text = cv2.imread ("D:\Python\message_yoroshiku.png")
holo = cv2.imread ("D:\Python\hologram_kira_sticker_color.png")
```

```python
white_min = np.array([255, 255, 255], np.uint8)
white_max = np.array([255, 255, 255], np.uint8)
```

```python
text_hsv = cv2.cvtColor(text, cv2.COLOR_BGR2HSV)
plt.imshow(cv2.cvtColor(text_hsv, cv2.COLOR_HSV2RGB))
```

```python
mask_white = cv2.inRange(text, white_min, white_max)
masked_text = cv2.bitwise_and(text,text, mask=mask_white)
plt.imshow(masked_text)
```

```python
holo_r = cv2.resize(holo, masked_text.shape[1::-1])
plt.imshow(holo_r)
```

```python
result = cv2.bitwise_and(masked_text, holo_r)
plt.imshow(result)
```
