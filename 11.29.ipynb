```python
import requests
from bs4 import BeautifulSoup
```

```python
load_url = "https://www.tokushima-u.ac.jp/recent/recent-list/"
html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
print(soup)
```

```python
print(soup.find("time").text)
print(soup.find("h2").text)
```

```python
pubtime_r=[]
for pubtime in soup.find_all("time")[0:10]:    
    print(pubtime.text)
    pubtime_r.append(pubtime.text)
print(pubtime_r)
```

```python
title_r=[]
for title in soup.find_all("h2")[0:10]: 
    print(title.text)
    title_r.append(title.text)
print(title_r)
```

```python
link_r=[]
links = soup.select("h2 a")
for link in links[0:10]:
  link = link.get("href")
  link_r.append(link)
  print(link)
print(link_r)
```

```python
links = soup.select("h2 a")
link_r = [link.get("href") for link in links[0:10]]
link_r
```

```python
import csv
import pandas as pd
```

```python
df = pd.DataFrame(zip(pubtime_r,title_r,link_r), columns = ["pubtime","title","link"])
df = df.set_index("pubtime")
df.to_csv("output_pd.csv")
df
```
